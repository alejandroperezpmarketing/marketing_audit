2. Elasticidad Precio-Demanda:

La elasticidad precio-demanda mide cómo responde la demanda ante cambios en el precio.
Fórmula:
Ep=%ΔDemanda%ΔPrecio
Ep​=%ΔPrecio%ΔDemanda​

    Si Ep>1Ep​>1: Demanda elástica (sensible al precio).
    Si Ep<1Ep​<1: Demanda inelástica (poco sensible al precio).

Ejemplo en Python: Cálculo de Elasticidad Precio-Demanda

import numpy as np

# Precios y demanda simulados
precios = np.array([5, 6, 7, 8, 9])
demandas = np.array([100, 90, 80, 70, 60])

# Cálculo de elasticidad
elasticidad = (np.diff(demandas) / demandas[:-1]) / (np.diff(precios) / precios[:-1])
print("Elasticidad:", elasticidad)

2. Simulación de Redes Logísticas Complejas en Python

Un enfoque avanzado para optimizar cadenas de suministro es simular múltiples rutas y determinar la más eficiente.

import networkx as nx

# Creación de una red logística compleja
G = nx.DiGraph()

# Añadir nodos (almacenes y tiendas)
nodos = ["Almacén", "Centro de Distribución", "Tienda A", "Tienda B", "Tienda C"]
G.add_nodes_from(nodos)

# Añadir conexiones con pesos (distancias en km)
G.add_weighted_edges_from([
    ("Almacén", "Centro de Distribución", 10),
    ("Centro de Distribución", "Tienda A", 15),
    ("Centro de Distribución", "Tienda B", 20),
    ("Centro de Distribución", "Tienda C", 25),
    ("Tienda A", "Tienda B", 5),
    ("Tienda B", "Tienda C", 10)
])

# Cálculo de la ruta más eficiente usando el algoritmo de Dijkstra
ruta_optima = nx.shortest_path(G, source="Almacén", target="Tienda C", weight="weight")
distancia_total = nx.shortest_path_length(G, source="Almacén", target="Tienda C", weight="weight")
print("Ruta óptima:", ruta_optima)
print("Distancia total:", distancia_total, "km")

1. Clasificación de Clientes en Marketing Digital: Algoritmo de Regresión Logística

El objetivo aquí es predecir si un cliente realizará una compra basada en características como edad, género y tiempo de visita al sitio web.
Paso 1: Preparar los datos

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Datos de ejemplo: Edad, Tiempo en el sitio (en minutos), Realizó compra (1: Sí, 0: No)
data = {
    'Edad': [25, 30, 35, 40, 22, 28, 34, 40, 55, 60],
    'Tiempo': [15, 25, 35, 30, 10, 50, 40, 60, 70, 20],
    'Compra': [1, 0, 1, 0, 1, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# Dividir datos en características (X) y etiquetas (y)
X = df[['Edad', 'Tiempo']]
y = df['Compra']

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalizar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

Paso 2: Crear el modelo de clasificación y evaluar

# Crear el modelo de regresión logística
model = LogisticRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Predicción
y_pred = model.predict(X_test)

# Evaluación
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Precisión: {accuracy * 100:.2f}%")
print("Matriz de confusión:")
print(conf_matrix)

Resultado esperado:

El modelo te indicará la probabilidad de que un cliente realice una compra basada en las características proporcionadas (edad y tiempo en el sitio).

from sklearn.cluster import KMeans

3. Optimización de Rutas para la Distribución de Productos (Algoritmo de Optimización)

Un aspecto clave del geomarketing es la optimización de rutas para minimizar costos y tiempo de entrega al seleccionar las rutas más eficientes para la distribución de productos a los clientes.
Ejemplo con Algoritmo Genético para Optimizar Rutas:

Usando un algoritmo genético, podemos encontrar la mejor ruta entre múltiples ubicaciones.

import random
import numpy as np
import matplotlib.pyplot as plt

# Definir las ubicaciones de las tiendas y clientes (coordenadas aleatorias)
ubicaciones = np.array([
    [0, 0],  # Almacén
    [4, 3],  # Cliente 1
    [5, 5],  # Cliente 2
    [6, 8],  # Cliente 3
    [7, 2]   # Cliente 4
])

# Calcular la distancia Euclidiana entre dos puntos
def distancia(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Crear la población inicial (individuos representan rutas)
def crear_ruta():
    ruta = list(range(1, len(ubicaciones)))
    random.shuffle(ruta)
    return [0] + ruta + [0]

# Función para calcular la longitud total de la ruta
def longitud_ruta(ruta):
    longitud = 0
    for i in range(len(ruta) - 1):
        longitud += distancia(ubicaciones[ruta[i]], ubicaciones[ruta[i + 1]])
    return longitud

# Algoritmo genético para optimizar la ruta
def algoritmo_genetico(generaciones=1000, poblacion_size=100, tasa_mutacion=0.1):
    poblacion = [crear_ruta() for _ in range(poblacion_size)]
    
    for generacion in range(generaciones):
        # Ordenar la población por la longitud de la ruta (menor es mejor)
        poblacion = sorted(poblacion, key=longitud_ruta)
        
        # Selección: mantener los mejores
        mejores = poblacion[:poblacion_size // 2]
        
        # Cruzamiento: crear nuevos individuos
        hijos = []
        for i in range(0, len(mejores), 2):
            padre1, padre2 = mejores[i], mejores[i+1]
            hijo = padre1[:len(padre1)//2] + [x for x in padre2 if x not in padre1[:len(padre1)//2]]
            hijos.append(hijo)
        
        # Mutación: cambiar algunas rutas
        for hijo in hijos:
            if random.random() < tasa_mutacion:
                i, j = random.sample(range(1, len(hijo) - 1), 2)
                hijo[i], hijo[j] = hijo[j], hijo[i]
        
        # Actualizar la población
        poblacion = mejores + hijos
    
    mejor_ruta = poblacion[0]
    return mejor_ruta, longitud_ruta(mejor_ruta)

# Optimizar ruta
mejor_ruta, mejor_longitud = algoritmo_genetico()
print(f"La mejor ruta es: {mejor_ruta}")
print(f"Longitud total de la ruta: {mejor_longitud:.2f} unidades")

# Visualizar la mejor ruta
ruta_coords = ubicaciones[mejor_ruta]
plt.plot(ruta_coords[:, 0], ruta_coords[:, 1], marker='o')
plt.title("Ruta Óptima de Distribución")
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
plt.show()

Resultado esperado:

El algoritmo encuentra la mejor ruta para visitar a los clientes, optimizando la distancia total recorrida, lo cual es crucial para reducir los costos logísticos y mejorar la eficiencia de la entrega.

La selección de nuevas ubicaciones de tienda es una de las aplicaciones más poderosas del geomarketing. Usando técnicas de machine learning, podemos analizar una serie de factores para identificar las ubicaciones más estratégicas para abrir nuevas tiendas.
Factores a considerar para la selección de ubicación:

    Proximidad a los clientes potenciales: La localización geográfica de los clientes es clave. Un análisis de proximidad y densidad poblacional ayudará a identificar las áreas con mayor demanda potencial.

    Competencia existente: Identificar cuántas tiendas competidoras existen en una zona ayudará a evaluar la saturación del mercado y determinar si hay espacio para una nueva tienda.

    Accesibilidad y tráfico: Las ubicaciones cerca de carreteras principales, zonas comerciales o áreas de alto tráfico son más atractivas para nuevos puntos de venta.

    Factores demográficos: Incluir datos como edad, ingresos, comportamiento de compra y estilo de vida de los consumidores en la zona.

Uso de Machine Learning para la selección de ubicaciones:

Podemos aplicar un algoritmo de regresión para predecir la rentabilidad de una nueva ubicación basada en diferentes variables. A continuación, se utiliza un ejemplo con un conjunto de datos ficticio.
Ejemplo en Python:

import numpy as np
import matplotlib.pyplot as plt

}
