```python
http = httplib2.Http()
```

# Extracción de head y body html de cada página

```python
resp_asiri, data_asiri = http.request(asiri_urls)
```

```python
# Codificando el html - body

asiri_html = data_asiri.decode('UTF-8')

# Limpieza de  codigo html

soup_asiri = BeautifulSoup(asiri_html)
```

```python
text = ' '.join(map(lambda p: p.text, soup_asiri.find_all(['p', 'span', 'h1', 'h2', 'h3', 'h4', 'li' 'ul'])))
text
```

```python
text.encode('ascii', errors='replace').replace(b"?", b" ")
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
constomStopWords = set(stopwords.words('spanish') + list(punctuation))

constomStopWords
```

```python
word_sent=[word for word in word_sent if word not in constomStopWords]
word_sent
```

```python
noWords = ['”', '“', '·', 'cookies', '«', 'asiri', '»','quieres', 'si', 'visión', 'to', '–','contacto',  'and', '©', '1','—',  '2','u', '7','are', '¿', '¿hasta', '¿por' , '100','¿y', '...', '``', '’', '48', '‘', '…' , 's.l', 'menú', '\r','\n' ]
freq_list = [word for word in word_sent if word not in noWords]
freq_list
```

```python
freq = FreqDist(freq_list)
freq
```

```python
nlargest(10, freq_list, key=freq.get)
```

```python
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
            
ranking
```

```python
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
